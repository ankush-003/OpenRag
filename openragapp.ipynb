```python
%%writefile app.py
from chainlit.server import app
from fastapi import Request, File, UploadFile
import os
import chainlit as cl
import requests
import random
from huggingface_hub import login
from llama_index.core import VectorStoreIndex,SimpleDirectoryReader,ServiceContext,PromptTemplate
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.llms.ollama import Ollama
from llama_index.core.settings import Settings
from llama_index.readers.notion import NotionPageReader
import torch
from llama_index.core.output_parsers import PydanticOutputParser
from typing import List
from transformers import BitsAndBytesConfig
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index.core import ServiceContext
from llama_index.embeddings.langchain import LangchainEmbedding
from llama_index.core.memory import ChatMemoryBuffer
from huggingface_hub import login
from chainlit.input_widget import TextInput, Slider
from pydantic import BaseModel
from llama_index.core.program import LLMTextCompletionProgram
from llama_index.core.output_parsers import ChainableOutputParser

class FlashCard(BaseModel):
    """Flashcard"""
    question: str
    ideal_ans: str
    user_ans: str
        
class Evaluation(BaseModel):
    """list of topics the user is weak in"""
    topics_user_is_weak_in: List[str]

class CustomEvaluationOutputParser(ChainableOutputParser):
    """Custom Evaluation output parser.
    Assume each subsequent line is the topic user is weak in.
    """

    def __init__(self, verbose: bool = False):
        self.verbose = verbose

    def parse(self, output: str) -> Evaluation:
        """Parse output."""
        if self.verbose:
            print(f"> Raw output: {output}")
        lines = output.split("\n")
        weak_topics = []
        for i in range(len(lines)):
            topics = lines[i].split(",")
            weak_topics.extend(topics)

        return Evaluation(topics_user_is_weak_in=weak_topics)

login(token=os.environ["HF_TOKEN"])

integration_token = os.environ["NOTION_TOKEN"]

system_prompt="""
You are a Q&A assistant. Your goal is to answer questions as
accurately as possible based on the instructions and context provided.
"""
## Default format supportable by LLama2
query_wrapper_prompt=PromptTemplate("<|USER|>{query_str}<|ASSISTANT|>")

docs = None
query_engine = None
chat_engine = None

# Loading LLM
llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    # uncomment this if using CUDA to reduce memory usage
    model_kwargs={"torch_dtype": torch.float16 , "load_in_4bit":True}
)   
# llm = Ollama(model='llama2', request_timeout=600.0)

embed_model=LangchainEmbedding(
    HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2"))

print("Models Loaded Successfully!")
    

@app.post('/upload/')
async def create_uploda_file(file: UploadFile = File(...)):
    print(file.filename)
    try:
        os.makedirs("data", exist_ok=True)
        path = f'data/{file.filename}'
        contents = await file.read()
        with open(path, 'wb') as f:
            f.write(contents)
        return {'filename': file.filename, 'path': path}    
    except Exception as e:
        return {'error': str(e)}

@cl.on_chat_start
async def on_chat_start():
    # setup
    cl.user_session.set("num_questions", 3)
    cl.user_session.set("token", 2)
    
    settings = await cl.ChatSettings(
        [
            Slider(
                id="num",
                label="Num Questions",
                initial=2,
                min=0,
                max=9,
                step=1,
            ),
            Slider(
                id="token",
                label="token",
                initial=1,
                min=1,
                max=2,
                step=1,
            ),
        ]
    ).send()
    Files = None

    while Files == None:
        Files = await cl.AskFileMessage(
            content = 'Please upload a file',
            max_size_mb=20,
            timeout=300,
            accept=['application/pdf']
        ).send()
    pdf_file = Files[0]
    
    with open(pdf_file.path, 'rb') as f:
        os.makedirs('data', exist_ok=True)
        with open(f'data/{pdf_file.name}', 'wb') as data_f:
            data_f.write(f.read())
            
    msg = cl.Message(content=f"Processing `{pdf_file.name}`...", disable_feedback=True)
    await msg.send()
    docs = SimpleDirectoryReader(input_files=[f'data/{pdf_file.name}']).load_data()
    await msg.remove()
    
    msg = cl.Message(content='Loading Models!', disable_feedback=True)
    await msg.send()
    Settings.llm = llm
    Settings.embed_model = embed_model
    Settings.chunk_size = 1024
    await msg.remove()
    
    cl.user_session.set("docs", docs)
    actions = [
        cl.Action(name='Ask a doubt', description='Ask a doubt!', value='Ask a doubt!'),
        cl.Action(name="Get flashcards", description="Get Flashcards!", value="Get Flashcards!"),
        cl.Action(name="Ask notion", description="Ask Notion", value="Ask Notion"),
    ]
    elements = [
      cl.Pdf(name=pdf_file.name, display="inline", path=f'data/{pdf_file.name}')
    ]
    await cl.Message(content=f'Start Interacting with `{pdf_file.name}`', actions=actions, disable_feedback=True, elements=elements).send()

@cl.action_callback('Ask a doubt')
async def act1():
    res = await cl.AskUserMessage(content="What is your doubt?", timeout=300).send()
    docs = cl.user_session.get("docs")
    index = VectorStoreIndex.from_documents(docs)
#     chat_engine = cl.user_session.get("chat_engine")
    chat_engine = index.as_query_engine()
    response = chat_engine.query(res['output'])
    elements = [
        cl.Text(name="response", content=response.response, display="inline")
    ]
    await cl.Message(content="Response to your query!", elements=elements).send()
    
@cl.on_settings_update
async def setup_agent(settings):
    cl.user_session.set("num_questions", settings["num"])
    cl.user_session.set("token", settings["token"])
    print("on_settings_update", settings)    

@cl.action_callback('Get flashcards')
async def act2():
    num_questions = cl.user_session.get("num_questions")
#     num_questions = 3
    token = cl.user_session.get("token")
    res = await cl.AskUserMessage(content="Which topic do you wanna learn today ?", timeout=300).send()

    topic = res["output"]
    docs = cl.user_session.get("docs")
    index = VectorStoreIndex.from_documents(docs)
    chat_engine = index.as_query_engine()
    
    if token == 1:
        verbose_level = "Make the explanation more detailed and give me a small example."
    elif token == 2:
        verbose_level = "Make the explanation brief and direct with the Keywords Highlighted."
    else:
        raise ValueError("Invalid token value. Token must be 1 or 2.")

    questions = [
        f"Define {topic}",
        f"Give a brief overview of {topic}",
        f"What is the primary concept of {topic}?",
        f"What are the key components of {topic}?",
        f"How is {topic} applied in real-world scenarios?",
        f"What are the common challenges associated with ?",
        f"Explain the process of {topic} in detail",
        f"How does {topic} contribute to specific fields or disciplines?",
        f"What are the implications of {topic} on a specific aspect?",
        f"Can you provide a case study or example of {topic} in action?"
    ]
    
    random.shuffle(questions)
    
    weak_topics = []
    
#     flashcards: List[FlashCard] = [FlashCard(question=q) for q in questions[:num_questions]]

    task_list = cl.TaskList()
    task_list.status = "Running..."
    
    
    for q in questions[:num_questions]:
        task = cl.Task(title=q, status=cl.TaskStatus.RUNNING)
        await task_list.add_task(task)
        await task_list.send()
        res = await cl.AskUserMessage(content=f"Question: {q}\nPlease enter your answer ✏️", timeout=300).send()
        msg = cl.Message(content=f"Analysing your answers...", disable_feedback=True)
        await msg.send()
    
        ideal_ans = chat_engine.query((q + verbose_level))
        msg.content = f"Expected answer:\n{ideal_ans}"
        await msg.update()
        
        prompt_template_str = f"""{ideal_ans.response}\nThis is the ideal answer for a question \
        Based on this answer {res["output"]} list the topics user is weak in by assessing both the ideal answer and given answer only, in the following format only
        with no additional text
        <topic user is weak in>, <topic user is weak in>
        the topics could go more than one line
        """
        program = LLMTextCompletionProgram.from_defaults(
            output_parser=CustomEvaluationOutputParser(verbose=True),
            output_cls=Evaluation,
            prompt_template_str=(prompt_template_str),
            verbose=True,
        )
        
        
        output = program()
#         print(output)
        weak_topics.extend(output.topics_user_is_weak_in)
#         print(weak_topics)

        task.status = cl.TaskStatus.DONE
        await task_list.send()
    
    actions = [
        cl.Action(name='Ask a doubt', description='Ask a doubt!', value='Ask a doubt!'),
        cl.Action(name="Get flashcards", description="Get Flashcards!", value="Get Flashcards!"),
        cl.Action(name="Ask notion", description="Ask Notion", value="Ask Notion"),
    ]
    
    res = llm.complete(f"After a test the user was founc to be weak in the following topics {weak_topics},  perform an analysis of the input abd list the most important topics the user is weak at")
    await cl.Message(content=f"Systems's Evaluation: {res}\n Please learn the following topics again!", actions=actions, disable_feedback=True).send()
    await task_list.send()
    
        

#     questions_to_query = questions[:num_questions]

#     answers = {}

#     for question in questions_to_query:
        
#         answers[question] = response

#     flashcards = []

#     for question, response in answers.items():
#         flashcards.append(f"Q: {question} \nA: {response}\n\n")
#     elements = [
#         cl.Text(name="flashcards", content="".join(flashcards), display="inline")
#     ]        
#     await cl.Message(content=f"Flashcard for Topic: {topic}\n\n", elements=elements).send()
    

@cl.on_message
async def main(message: cl.Message):
    user_ans = f"Received: {message.content}"
    
    
@cl.action_callback('Ask notion')
async def act3():
    res = await cl.AskUserMessage(content="Input your page id", timeout=300).send()
    notion_id = res['output']
    part1 = notion_id[:8]
    part2 = notion_id[8:12]
    part3 = notion_id[12:16]
    part4 = notion_id[16:20]
    part5 = notion_id[20:]
    
    # Combine the parts with hyphens
    formatted_id = f"{part1}-{part2}-{part3}-{part4}-{part5}"
    docs = NotionPageReader(integration_token=integration_token).load_data(
        page_ids=[formatted_id]
    )
    index = VectorStoreIndex.from_documents(docs)
#     chat_engine = cl.user_session.get("chat_engine")
    chat_engine = index.as_query_engine()
    res = await cl.AskUserMessage(content="What do you wanna know ?", timeout=120).send()
    response = chat_engine.query(res['output'])
    elements = [
        cl.Text(name="response", content=response.response, display="inline")
    ]   
    await cl.Message(content="Answer to your query!", elements=elements).send()
```
